```python
from transformers import pipeline
```

```python
unmasker = pipeline('fill-mask', model='bert-base-multilingual-cased')
```

```python
unmasker_ru = pipeline('fill-mask', model="DeepPavlov/rubert-base-cased")
```

```python
with open('citytest.txt', encoding='utf-8') as f:
    data = f.readlines()
```

```python
for d in data:
    res = unmasker(d)
    sent = d
    topic = 'cities'
    masked = 'hyperonym'
    expect = 'города'
    pred1 = res[0]['token_str']
    pred2 = res[1]['token_str']
    pred3 = res[2]['token_str']
    pred4 = res[3]['token_str']
    pred5 = res[4]['token_str']
    cities_df.loc[len(cities_df.index )] = [sent, topic, masked, expect, pred1, pred2, pred3, pred4, pred5]
    
    res_ru = unmasker_ru(d)
    pred1 = res_ru[0]['token_str']
    pred2 = res_ru[1]['token_str']
    pred3 = res_ru[2]['token_str']
    pred4 = res_ru[3]['token_str']
    pred5 = res_ru[4]['token_str']
    cities_df_ru.loc[len(cities_df_ru.index )] = [sent, topic, masked, expect, pred1, pred2, pred3, pred4, pred5]
    
```
